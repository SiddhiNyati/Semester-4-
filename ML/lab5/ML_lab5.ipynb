#**Data Cleaning:**

```python
cleaned_df = df_train.drop(columns=["Posted On", "Area Type", "Tenant Preferred", "Point of Contact"])
cleaned_df
```

```python
missing_values = cleaned_df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
missing_values = cleaned_df.isna().sum()
print("Missing values:\n", missing_values)
```

No null or NA values

```python
categorical_columns = cleaned_df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
for column in categorical_columns:
    cleaned_df[column] = label_encoder.fit_transform(cleaned_df[column])

cleaned_df
```

# **Regression:**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train_part = cleaned_df[['Rent']]
train_part
X_val = df_val[['Rent']]
y_val = df_val['Size']
model = LinearRegression()
model.fit(train_part,cleaned_df['Size'])
pred = model.predict(X_val)
mse = mean_squared_error(y_val,pred)
print("Mean Squared Error on Validation Set:", mse)
```

```python
import numpy as np
```

```python
def calculate_rmse(predictions, actual_values):
    sq_errors = (predictions - actual_values) ** 2
    sum_sq_errors = np.sum(sq_errors)
    N = len(actual_values)
    mean_sq_errors = sum_sq_errors / N
    rmse = np.sqrt(mean_sq_errors)

    return rmse
```

```python
rmse = calculate_rmse(pred, y_val)
print("Root Mean Squared Error (RMSE):", rmse)
```
