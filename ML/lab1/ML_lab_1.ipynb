**Q6.  Split the dataset into the following partitions. Note that the partitions should not have any missing values and all the columns should have a common datatype.**


**Training set: All movies with release year <=2010**

**Test Set: All movies with release year > 2010**

```python
dataset.dropna(inplace=True)


dataset['year'] = dataset['year'].astype(int)

common_datatype = dataset.dtypes.unique()[0]

data = dataset.astype(common_datatype)

# defining training & test set:-
training_set = data[dataset['year'] <= 2010]
test_set = data[dataset['year'] > 2010]
```

**Q8.  Do a random split of the training set into 80% training and the remaining 20% validation set (approximately) based on the unique movies, i.e., 80% of the movies in the with be randomly separated to train the models.**

```python
unique_movies = training_set['title'].unique()
train_size = int(0.8 * len(unique_movies))
train_movies = unique_movies[:train_size]
validation_movies = unique_movies[train_size:]
train_set = training_set[training_set['title'].isin(train_movies)]
validation_set = training_set[training_set['title'].isin(validation_movies)]
```

# **Plotting ( Use both Maplotlib and Seaborn ):**

**Q1. Create a bar plot which shows the number of characters for a particular rating in the training set. The x axis has the ratings and Y- the axis has the count of characters.**

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(80, 20))
sns.countplot(x='n', data=train_set)
plt.xlabel('Rating')
plt.ylabel('Number of Characters')
plt.title('Number of Characters for Each Rating in the Training Set')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

**Q2. Draw a time series line plot, with years on the x-axis, and the average rating of the actors/actresses on the y-axis. This will be to see if there has been a pattern in the likeness or dislikeness of movies over the years.**

```python
average_rating_by_year = data.groupby('year')['n'].mean()

plt.plot(average_rating_by_year.index, average_rating_by_year.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Rating of Actors/Actresses Over the Years')
plt.grid(True)
plt.show()
```
