**Q5. Display the average stock prices for the different years and for all the 4 stocks. Plot it on a graph for comparison with other companies. Create an appropriate legend by keeping different stocks represented by different coloured lines.**

Code for displaying the average stock prices for the different years and all the 4 stocks:-

```python
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year

avg = df.groupby('year').mean()

print("Average stock prices for different years:")
avg
```

Code to plot it on a graph for comparison with other companies:-

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))

for column in avg.columns:
    plt.plot(avg.index, avg[column], label=column)

plt.title('Average Stock Prices for Different Years')
plt.xlabel('Year')
plt.ylabel('Average Stock Price')
plt.legend()
plt.show()
```

**Q6. Compute moving average values (3 day moving average and 10 day moving average) for the different stocks. Create separate columns for each stock and moving average. Concatenate the new columns with the original dataframe after naming the columns. For 4 initial stocks, the resultant dataframe will have 8 new columns plus the original columns.**

```python
for company in ['AA', 'GE', 'IBM', 'MSFT']:
    df[f'3day_{company}'] = df[company].rolling(window=3).mean()
    df[f'10day_{company}'] = df[company].rolling(window=10).mean()

df
```

**Q7. Find the average stock prices for each month for IBM and plot the trend between the years 1995 – 2010. Compare the trend with the Microsoft Corporation stocks (MSFT). Create appropriate legend and axis labels.**

```python
df1 = df[(df['date'].dt.year >= 1995) & (df['date'].dt.year <= 2010)]

df1['year_month'] = df1['date'].dt.to_period('M')

#calculate the average for IBM and MSFT
avg_month = df1.groupby(['year_month']).mean()[['IBM', 'MSFT']]

avg_month
```

```python
plt.figure(figsize=(40, 20))
plt.plot(avg_month.index.astype(str), avg_month['IBM'], label='IBM')
plt.plot(avg_month.index.astype(str), avg_month['MSFT'], label='MSFT')

plt.xlabel('Year-Month')
plt.ylabel('Average Stock Price')
plt.title('Average Stock Prices for IBM and MSFT (1995-2010)')
plt.legend()

plt.xticks(rotation=45)
plt.show()
```

**Q8. Plot the moving average values with actual values for the IBM stock for the year 2010. Show the moving averages as dashed line with appropriate labels and legend.**

```python
df_2010 = df[df['date'].dt.year == 2010]

df_2010['IBM_3day_MA'] = df_2010['IBM'].rolling(window=3, min_periods=1).mean()
df_2010['IBM_10day_MA'] = df_2010['IBM'].rolling(window=10, min_periods=1).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_2010['date'], df_2010['IBM'], label='IBM - Actual Values')
plt.plot(df_2010['date'], df_2010['IBM_3day_MA'], '--', label='IBM - 3-day Moving Average')
plt.plot(df_2010['date'], df_2010['IBM_10day_MA'], '--', label='IBM - 10-day Moving Average')


plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('IBM Stock Prices and Moving Averages (Year 2010)')
plt.legend()


plt.xticks(rotation=45)
plt.show()
```
