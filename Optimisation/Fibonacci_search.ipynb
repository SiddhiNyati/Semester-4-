```python
def f(x):
    return x**2 - 6*x + 15

def fibonacci(n):
    fib = [1, 1]
    for i in range(2, n+1):
        fib.append(fib[i-1] + fib[i-2])
    return fib

def fibonacci_search(A, B, n):
    iteration = 1

    while n >= 1:
        fib = fibonacci(n)
        x1 = A + (fib[n-2] / fib[n]) * (B - A)
        x2 = B - (fib[n-2] / fib[n]) * (B - A)

        fx1 = f(x1)
        fx2 = f(x2)

        print(f"Iteration {iteration}:")
        print(f"A = {A}, B = {B}, x1 = {x1}, x2 = {x2}")
        print(f"f(x1) = {fx1}, f(x2) = {fx2}")
        print(f"n = {n}")

        if fx1 > fx2:
            A = x1
            x1 = x2
            B = B
            x2 = B - (fib[n-3] / fib[n]) * (B - A)
        else:
            B = x2
            x2 = x1
            A = A
            x1 = A + (fib[n-3] / fib[n]) * (B - A)

        n -= 1
        iteration += 1

    return (A + B) / 2

# Initial interval
A = 2
B = 5
n = 28  # Number of iterations

# Perform Fibonacci search
result = fibonacci_search(A, B, n)
print("Final result:", result)
```
