```python
def f(x):
    return x**2 - 6*x + 15

def golden_section_search(A, B, epsilon=0.00001):
    iteration = 1
    while B - A > epsilon:
        x1 = 0.618 * (B - A) + A
        x2 = 0.382 * (B - A) + A

        fx1 = f(x1)
        fx2 = f(x2)

        print(f"Iteration {iteration}:")
        print(f"A = {round(A, 5)}, B = {round(B, 5)}, x1 = {round(x1, 5)}, x2 = {round(x2, 5)}")
        print(f"f(x1) = {round(fx1, 5)}, f(x2) = {round(fx2, 5)}")

        if fx1 > fx2:
            B = x1
            A = A
            x1 = x2
            x2 = 0.382*(B-A) + A
        else:
            A = x2
            B = B
            x2 = x1
            x1 = 0.618*(B-A) + A

        iteration += 1

    return round((A + B) / 2, 5)

# Initial interval
A = 0
B = 10

# Perform golden section search
result = golden_section_search(A, B)
print("Final result:", result)
```
